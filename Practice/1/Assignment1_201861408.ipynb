# Answer 5
Percentage of males and females that are fan of star wars.

```python
df['gender'].value_counts()
```

```python
male = df[(df['gender'] == 'Male')]
male_fan = male[(male['fan'] == 'Yes')]
percentage_of_male_fan = (len(male_fan) / len(df)) * 100
percentage_of_male_fan
```

```python
female = df[(df['gender'] == 'Female')]
female_fan = female[(female['fan'] == 'Yes')]
percentage_of_female_fan = (len(female_fan) / len(df)) * 100
percentage_of_female_fan
```
