```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('population.csv')
df
```

# Answer 1

```python
country_df = df[df['Country'] == 'Canada']
country_df
```

```python
x = country_df['Year']
y = country_df['Population']
plt.xlabel('Year')
plt.ylabel('Population')
plt.title("Canada's Population Growth")
plt.yticks([1.5e7, 2.0e7, 2.5e7, 3.0e7, 3.5e7], ['15M', '20M', '25M', '30M', '35M'])
plt.plot(x,y)
plt.show()
```

# Answer 2

```python
df['Year'].unique()
```

```python
year = 1955
percentage_list = []
while year <= 2010:
    urban_df = df[df['Year'] == year]
    global_population = urban_df['Population'].sum()
    urban_population = urban_df['UrbanPopulation'].sum()
    percentage = (urban_population / global_population) * 100
    percentage_list.append(percentage)
    year = year + 5
while year <= 2020:
    urban_df = df[df['Year'] == year]
    global_population = urban_df['Population'].sum()
    urban_population = urban_df['UrbanPopulation'].sum()
    percentage = (urban_population / global_population) * 100
    percentage_list.append(percentage)
    year = year + 1
```

```python
x1 = [1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2016, 2017, 2018, 2019, 2020]
y1 = percentage_list
plt.xlabel('Year')
plt.ylabel('Urban Percentage')
plt.title('Percentage of Global Population living in Urban Areas')
plt.ylim(30,57.5)
plt.yticks([30, 35, 40, 45, 50, 55], ['30%', '35%', '40%', '45%', '50%', '55%'])
plt.plot(x1, y1)
plt.show()
```

# Answer 3

```python
df['Continent'].unique()
```

```python
#fig = plt.Figure()
#ax = fig.axes([0,0,1,1])
fig, ax = plt.subplots(figsize=(8, 5))
continent = ['Africa', 'Asia', 'Europe', 'Latin America and the Caribbean', 'North America', 'Oceania']
i = 0

while i < len(continent):
    year = 1955
    population_list = []
    continent_df = df[df['Continent'] == continent[i]]
    while year <= 2010:
        population_df = continent_df[continent_df['Year'] == year]
        population = population_df['YearlyChange'].sum()
        population_list.append(population)
        year = year + 5
    while year <= 2020:
        population_df = continent_df[continent_df['Year'] == year]
        population = population_df['YearlyChange'].sum()
        population_list.append(population)
        year = year + 1
    ax.plot(x1, population_list, label = continent[i])
    i = i + 1

#plt.ylim(,6.5e7)
ax.set_xlabel('Year')
ax.set_ylabel('Yearly Population Change')
ax.set_title('Yearly Population Change by Continent')
ax.legend(loc = 'center', frameon = False)
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(8,5))
year_range_df = df[(df['Year'] == 2019) | (df['Year'] == 2020)]
i = 0
while i < len(continent):
    median_and_percent_df = year_range_df[year_range_df['Continent'] == continent[i]]
    ax.scatter(median_and_percent_df['MedianAge'], median_and_percent_df['Yearly%Change'], label = continent[i])
    i = i + 1
ax.set_xlabel('Median Age')
ax.set_ylabel('Percent')
ax.set_title('Country Population Growths (2019 to 2020)')
ax.set_xlim(14,50)
ax.set_ylim(-2.9, 4.2)
ax.legend(loc = 'lower left', frameon = False)
fig.tight_layout()
plt.show()
```
