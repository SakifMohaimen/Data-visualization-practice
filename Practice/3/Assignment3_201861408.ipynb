```python
year = [1992, 2012, 1920, 1900, 1932, 1952, 2000, 1996, 1912, 1924, 1948,
       2008, 2016, 2004, 1984, 1968, 1972, 1988, 1936, 1956, 1960, 1928,
       1976, 1980, 1964, 1906, 1904, 1908, 1896]
year.sort()
```

```python
year
```

```python
new_df['Sex'].unique()
```

```python
male_df = new_df[new_df['Sex'] == 'M']
male_df
```

```python
male_df['Sex'].unique()
```

```python
male_df = male_df.drop(columns = ['ID', 'Sex', 'Age', 'Weight', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event', 'Medal'])
male_df
```

```python
male_df = male_df.dropna()
male_df
```

```python
female_df = new_df[new_df['Sex'] == 'F']
female_df
```

```python
female_df['Sex'].unique()
```

```python
female_df = female_df.drop(columns = ['ID', 'Sex', 'Age', 'Weight', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event', 'Medal'])
female_df
```

```python
female_df = female_df.dropna()
female_df
```

```python
i = 0
avg_male_height_list = []
avg_female_height_list = []

male_upper_std_list = []
male_lower_std_list = []


female_upper_std_list = []
female_lower_std_list = []
while i < len(year):
    new_male_df = male_df[male_df['Year'] == year[i]]
    new_male_df = new_male_df.drop_duplicates(['Name'])
    avg_male_height = new_male_df['Height'].mean()
    avg_male_std = new_male_df['Height'].std()
    avg_male_height_list.append(avg_male_height)
    male_upper = avg_male_height + avg_male_std
    male_lower = avg_male_height - avg_male_std
    male_upper_std_list.append(male_upper)
    male_lower_std_list.append(male_lower)
    
    
    new_female_df = female_df[female_df['Year'] == year[i]]
    new_female_df = new_female_df.drop_duplicates(['Name'])
    avg_female_height = new_female_df['Height'].mean()
    avg_female_std = new_female_df['Height'].std()
    avg_female_height_list.append(avg_female_height)
    female_upper = avg_female_height + avg_female_std
    female_lower = avg_female_height - avg_female_std
    female_upper_std_list.append(female_upper)
    female_lower_std_list.append(female_lower)
    i = i + 1
```

```python
avg_male_height_list
```

```python
avg_female_height_list
```

```python
fig, ax = plt.subplots(figsize = (7,5))
x = year
y1 = avg_male_height_list
y2 = avg_female_height_list

m_up = male_upper_std_list
m_low = male_lower_std_list

f_up = female_upper_std_list
f_low = female_lower_std_list


ax.plot(x, y1, color = 'blue', label = 'Male')
ax.plot(x, y2, color = 'red', label = 'Female')
ax.set_ylim(140,200)
ax.set_xlabel('Year')
ax.set_ylabel('Height (cm)')
ax.set_title('Average Olympian Height (+/- std dev) in the Summer Olympics')

plt.fill_between(x, m_up, m_low, alpha = 0.2)
plt.fill_between(x, f_up, f_low, alpha = 0.2)
plt.legend(loc = 'lower left')
plt.show()
```
