```python
import pandas as pd
import matplotlib. pyplot as plt
import seaborn as sns

df_p = pd.read_csv('population.csv')
df_p
```

```python
df_p.info()
```

```python
df_p['Continent'].unique()
```

```python
df_p.sort_values(by = 'Continent', inplace = True)
df_p
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax = sns.lineplot(x = 'Year', y = 'YearlyChange', data = df_p, hue = 'Continent', ci = None, legend = True)
ax.legend(loc = 'center', frameon = False)
#ax.set_ylim(0,6.5e7)
ax.set_yticks([0e6, 0.2e6, 0.4e6, 0.6e6, 0.8e6, 1.0e6, 1.2e6], ['0', '1', '2', '3', '4', '5', '6'])
ax.set_xlabel('Year')
ax.set_ylabel('Yearly Population Change')
ax.set_title('Yearly Population Change by Continent')
plt.show()
```

```python
n = df_p[df_p['Continent'] == 'North America']
n['Year'].unique()
```

```python
year = [2017, 2016, 2018, 2019, 1960, 1965, 1970, 1975, 1980, 1985, 1990,
       1995, 2015, 2000, 2010, 2020, 2005, 1955]
year.sort()
year
```

```python
i = 0
sum1 = 0
sumL = []
while i < len(year):
    north = n[n['Year'] == year[i]]
    sum1 = north['YearlyChange'].sum()
    sumL.append(sum1)
    i = i + 1
sumL
```

```python
olympic_df = pd.read_csv('olympic_athletes.csv')
olympic_df
```

```python
olympic_df['Season'].unique()
```

```python
olympic_df = olympic_df[olympic_df['Season'] == 'Summer']
olympic_df
```

```python
olympic_df = olympic_df.drop(columns = ['ID', 'Age', 'Weight', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event', 'Medal'])
```

```python
olympic_df.sort_values(by = 'Year', inplace = True)
```

```python
olympic_df.dropna()
```

```python
olympic_df = olympic_df.drop_duplicates(subset = ['Name', 'Year'], keep = 'first')
olympic_df
```

```python
olympic_df.dropna(inplace = True)
```

```python
olympic_df
```

```python
df_avg_height = olympic_df.groupby(['Year', 'Sex'])['Height'].agg(['mean', 'std']).reset_index()
#df_avg_height['std_up'] = df_avg_height['mean'] + df_avg_height['std']
#df_avg_height['std_low'] = df_avg_height['mean'] - df_avg_height['std']
#m = df_avg_height[df_avg_height['Sex'] == 'M']
#f = df_avg_height[df_avg_height['Sex'] == 'F']
df_avg_height
```

```python
colours = ['blue', 'red']
fig, ax = plt.subplots(figsize = (7,5))
ax = sns.lineplot(data=df_avg_height, x='Year', y='mean', hue='Sex', ci = 'sd', palette = colours, legend = True)
ax.set_title('Average Olympian Height (+/- std dev) in the Summer Olympics')
ax.set_ylabel('Height (cm)')
ax.set_ylim(140,200)
#sns.set(style="darkgrid")
plt.legend(loc = 'lower left')
plt.show()
```

```python
game_df = pd.read_csv('game_consoles.csv')
game_df = game_df.drop(columns = ['Type', 'Release'])
game_df
```

```python
group = game_df.groupby(['Company', 'Console'])


inner_colors = ['#156EAF', '#DB2018', '#56B45B']
outer_colors = ['#5599CC', '#EA6727', '#83C143']


fig, ax = plt.subplots(figsize=(10, 10))


sizes = group['Units_sold'].sum()

ax.pie(sizes, labels=sizes.index.get_level_values(1), colors=outer_colors, startangle=343,
       wedgeprops=dict(width=0.4, edgecolor='w'))
plt.title('Best Selling Video Game Consoles')
plt.show()
```
