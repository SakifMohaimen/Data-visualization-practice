```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets

df = pd.read_csv('names.csv')

colors = ['#206095', '#004662', '#A8BD3A', '#27A0CC', '#118C7B', '#F66068']

def main_plot(names, gender):
    
    # new_df is based on the given gender
    new_df = df[df['sex'] == gender]
    

    sns.set_style("whitegrid")
    fig, ax = plt.subplots(figsize=(10, 8))
    

    # top 100 names in year 2020
    top_100_df = new_df[new_df['year'] == 2020].head(100)
    top_100_name = top_100_df['name'].str.strip()
    for name in top_100_name:
        
        ax.plot(new_df[new_df['name'].str.strip() == name]['year'],
                new_df[new_df['name'].str.strip() == name]['rank'],
                color='lightgrey', alpha=0.5)
        
    df2 = new_df
    df2['name'] = new_df['name'].str.strip()
    

    # name_str will hold the names that the user inputs and will pass them to new_list based on '' space
    name_str = ''
    new_list = []
    for name in names:
        if name != '':
            name_str = name_str + name
        else:
            name_str = name_str + " " + name
    
    new_list = name_str.split()
    
    if len(new_list) > 0 and len(new_list) <= 6:
        i = len(new_list)
        # this if condition means the plot will highlight only 6 names
        if i <= 6 :
            j = 0
            while j < len(new_list):
                print(new_list[j])
                # For understanding the whole plot easily I am printing all the name. It will print max 6 names as per
                # the requirement. If the user provides another name then it won't print and thus won't highlight the name
                # That is why I am printing the names on which the plot will be highlighted
                # converting the name to uppercase will maintain name insensitivity
                ax.plot(df2[df2['name'] == new_list[j].upper()]['year'],
                        df2[df2['name'] == new_list[j].upper()]['rank'],
                        color=colors[j], linewidth=2, label = new_list[j].upper())
                ax.scatter(df2[df2['name'] == new_list[j].upper()]['year'],
                           df2[df2['name'] == new_list[j].upper()]['rank'],
                           color = colors[j])
                ax.legend(loc = 'lower right')
                j = j + 1
            i = i + 1
    
    ax.invert_yaxis()
    ax.set_title('Popularity ranking(1 being the most popular)')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_yticks([1, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
    ax.set_xticks(range(1996, 2021, 2))
    
    ax.grid(axis='x')
    sns.despine(left=True, right=True, top=True)
    
#widgets
names_box = widgets.Text(value='', placeholder='Enter names separated by space', description='Names:', disabled = False)
buttons = widgets.ToggleButtons(options=['boy', 'girl'], description = '', button_style='info')
final_plot = widgets.interactive_output(main_plot, {'names': names_box, 'gender': buttons})

display(widgets.VBox([buttons, names_box]))
display(final_plot)

## Note - the plot takes some time to show all the lines of the name
```
