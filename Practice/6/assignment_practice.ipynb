```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('students.csv')
df
```

```python
canada = gpd.read_file('canada.shp')
canada.plot()
```

```python
canada
```

```python
df['percentage'] = df['Total students'] / df['Population'] * 100
df
```

```python
canada = canada.merge(df, on = 'Province', how = 'left')
canada
```

```python
canada = canada.fillna(0)
canada
```

```python
fig, ax = plt.subplots(figsize=(10,8))
canada.plot(column = 'percentage', cmap = 'Reds', linewidth=0.5, edgecolor='black', ax=ax)
ax.set_title('Students as Percentage of Population')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
#ax.set_xticks([0, 4, 5, 6, 7, 8, 9], ['', '', '', '', '', '', ''])
#ax.set_yticks([0, 1, 2, 3, 4, 5], ['', '', '', '', '', ''])
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
fuel_df = pd.read_csv('fuel-efficiency.csv')
fuel_df
```

```python
import plotly.express as px
months = ['January', 'February', 'March', 'April', 'May', 'June',
         'July', 'August', 'September', 'October', 'November', 'December']
fuel_df['month'] = fuel_df['month'].apply(lambda x: months[x-1])
px.line_polar(r='efficiency', theta=months, data_frame=fuel_df, line_close=True, line_shape='spline', range_r = [0, 12])
```
